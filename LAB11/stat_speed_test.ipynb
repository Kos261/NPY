```python
import numpy as np

sample = np.genfromtxt('sample.csv', delimiter=' ')
#print(sample)
print(type(sample))
print(np.shape(sample))


%time np.mean(sample, axis=0)
%time np.median(sample, axis=0)
%time np.std(sample, axis=0)
```

```python
from statistics import mean, median, stdev
import csv

data = []
with open("sample.csv", newline='') as f:
    reader = csv.reader(f, delimiter=' ')
    for row in reader:
        # zamień wszystkie wartości tekstowe na float
        float_row = [float(value) for value in row]
        data.append(float_row)

print(data)
```
